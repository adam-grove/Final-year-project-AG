# Data Formatting 
The output from XCMS only needs columns renaming

```python


def rename_col(col_name):
  match = re.match(r"Combined_P(\d+)_t(\d+)h\.mzML", col_name)
  if match:
    # Use captured groups for renaming
    return f"P{match.group(1)}_t{match.group(2)}h"
  else:
    return ""  # Keep original name for non-matching columns (optional)

matching_cols = any(re.match(r"Combined_P(\d+)_t(\d+)h\.mzML", col) for col in df.columns)

if matching_cols:
  # Apply renaming function if there are matching columns
  df.columns = df.columns.map(rename_col)
else:
  print("No columns match the pattern - Probably the result of only running this cell")


df.to_csv("XCMS_Feature_Matrix.csv", index=False)


df
```
